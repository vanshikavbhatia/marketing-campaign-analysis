```python
import pandas as pd

# Load your data (adjust path if needed)
df = pd.read_csv('marketing_campaign.csv')

# Calculate CTR (clicks divided by impressions)
df['ctr'] = df['clicks'] / df['impressions']

# Group by channel_name and calculate average CTR
ctr_by_channel = df.groupby('channel_name')['ctr'].mean().sort_values(ascending=False)

# Display the results
print(ctr_by_channel)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index for easier plotting
ctr_by_channel = ctr_by_channel.reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=ctr_by_channel, x='channel_name', y='ctr', palette='viridis')
plt.xticks(rotation=45)
plt.title('Average Click-Through Rate (CTR) by Marketing Channel')
plt.xlabel('Channel')
plt.ylabel('Average CTR')
plt.tight_layout()
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("marketing_campaign.csv")
df['ctr'] = df['clicks'] / df['impressions']

st.title("üìä Marketing Campaign Dashboard")

channel = st.sidebar.selectbox("Choose a Channel", df['channel_name'].unique())
filtered_df = df[df['channel_name'] == channel]

st.subheader(f"üìå Performance of '{channel}' Channel")
st.metric("‚úÖ Avg CTR", round(filtered_df['ctr'].mean(), 4))
st.metric("üéØ Total Clicks", int(filtered_df['clicks'].sum()))
st.metric("üëÅÔ∏è Total Impressions", int(filtered_df['impressions'].sum()))
```
